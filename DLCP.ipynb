```python
import os
import cv2
import numpy as np

iam_path = "C:/Users/Dell/Desktop/SaBaHaT/SEMESTER 6/DL/Project/words"

# Getting the list of all image files in the dataset
image_files = []
for subdir, _, files in os.walk(iam_path):
    for file in files:
        if file.endswith('.png'):
            image_files.append(os.path.join(subdir, file))

# Loading the images and labels
images = []
labels = []
for image_file in image_files:
    # Extract the label from the file name
    label = image_file.split('/')[-1].split('-')[1][0]

    # Load the image and convert it to grayscale
    image = cv2.imread(image_file, cv2.IMREAD_GRAYSCALE)
    # Check the dimensions of the input image
    print('Input image dim:', image.shape)
    # Resize the image 
    image = cv2.resize(image, (28, 28))
    # Check the dimensions of the output image
    print('Output image dim:', image.shape)
    # Add the image and label to the lists
    images.append(image)
    labels.append(label)#ord(label) - 97)  # Convert the label to a numerical value (0-25)


# Converting the lists to numpy arrays
images = np.array(images)
labels = np.array(labels)

```

```python
# Saving the npy images and labels
np.save('iam_handwriting_data.npy', images)
np.save('iam_handwriting_labels.npy', labels)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import tensorflow as tf

# Loading dataset
data = np.load('iam_handwriting_data.npy')
labels = np.load('iam_handwriting_labels.npy')

# Splitting the dataset into training and testing 
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2)

X_train = X_train.reshape(X_train.shape[0], X_train.shape[1] * X_train.shape[2])
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1] * X_test.shape[2])

# Using KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Evaluating the classifier
accuracy = knn.score(X_test, y_test)
print('Accuracy:', accuracy)
```

```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow import keras

# Load the training and validation data
with open("C:/Users/Dell/Desktop/SaBaHaT/SEMESTER 6/DL/Project/the.txt") as f:
    train_data = f.readlines()
train_data = [line.strip().split(" ") for line in train_data]
with open("C:/Users/Dell/Desktop/SaBaHaT/SEMESTER 6/DL/Project/the.txt") as f:
    val_data = f.readlines()
val_data = [line.strip().split(" ") for line in val_data]

# Define a dictionary mapping characters to integers
char_to_num = {}
for line in train_data + val_data:
    for char in line[8:]:
        if char not in char_to_num:
            char_to_num[char] = len(char_to_num)

# function to load and preprocess an image
def preprocess_image(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (256, 56))
    img = img.astype(np.float32) / 255.0
    img = np.expand_dims(img, axis=-1)
    return img

img_h=56
img_w=256
num_classes = len(char_to_num)

# Define the model architecture
model = keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), activation="relu", input_shape=(56, 256, 1)),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(64, (3, 3), activation="relu"),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Conv2D(128, (3, 3), activation="relu"),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(16*256, activation="relu"),
    keras.layers.Reshape((16, 256)),
    keras.layers.Bidirectional(keras.layers.LSTM(128, return_sequences=True)),
    keras.layers.Bidirectional(keras.layers.LSTM(64)),
    keras.layers.Dense(num_classes, activation="softmax")
])

# Compile the model
optimizer = keras.optimizers.Adam()
model.compile(loss="sparse_categorical_crossentropy", 
              optimizer=optimizer, metrics=["accuracy"])


# Define the training and validation datasets
train_dataset = tf.data.Dataset.from_generator(
    lambda: ((preprocess_image("C:/Users/Dell/Desktop/SaBaHaT/SEMESTER 6/DL/Project/the.png"), [char_to_num[x] for x in line[8:]]) for line in train_data),
    output_types=(tf.float32, tf.int32),
    output_shapes=((56, 256, 1), (None,))
).batch(128).prefetch(tf.data.experimental.AUTOTUNE)

val_dataset = tf.data.Dataset.from_generator(
    lambda: ((preprocess_image("C:/Users/Dell/Desktop/SaBaHaT/SEMESTER 6/DL/Project/the.png"), [char_to_num[x] for x in line[8:]]) for line in val_data),
output_types=(tf.float32, tf.int32),
output_shapes=((56, 256, 1), (None,))
).batch(128).prefetch(tf.data.experimental.AUTOTUNE)
```

```python
history = model.fit(train_dataset, 
                    epochs=25, 
                    validation_data=val_dataset)
```

```python
# Plotting Accuracy vs Epoch plot
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
# Plotting Loss vs Epoch plot
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
